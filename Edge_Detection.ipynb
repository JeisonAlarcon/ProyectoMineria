```python
import os
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
Path = "/content/drive/MyDrive/Colab Notebooks/DATA/PokemonDataset/"
# Lista con todos los nombre del fichero
Images_URLS = os.listdir(Path + "OutputPNG")
Images_URLS[:10]
```

```python
# https://programarfacil.com/blog/vision-artificial/detector-de-bordes-canny-opencv/
for i in Images_URLS:
  Original = cv2.imread(Path + "OutputPNG/" + i)
  #  ■ Convertimos a escala de grises
  Gris = cv2.cvtColor(Original, cv2.COLOR_BGR2GRAY)
  #  ■ Aplicar suavizado Gaussiano
  Gauus  = cv2.GaussianBlur(Gris, (5,5), 0)
  Bordes = cv2.Canny(Gauus, 100, 250) # image, minVal, maxVal
  #  ■ Invirtiendo los colores (negro -> blanco)
  #    » Ojo la imagen resultante es (., ., 1) y necesitamos que sea (., ., 3)
  InvCol = cv2.cvtColor(cv2.bitwise_not(Bordes), cv2.COLOR_GRAY2BGR)
  # print(Original.shape, InvCol.shape)

  plt.subplot(121), plt.imshow(Original)
  plt.title("Imagen Original"), plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(InvCol, cmap = "gray")
  plt.title("Imagen Bordeada"), plt.xticks([]), plt.yticks([])
  plt.show()

  #  ■ Guardando la imagen generada
  cv2.imwrite(Path + "Input/" + i[0:3] + ".jpg", InvCol)
```
